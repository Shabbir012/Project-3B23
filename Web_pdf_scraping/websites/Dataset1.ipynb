```python
#import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
import nltk

nltk.download('punkt')
```

```python
#input the link to the website to extract its contents
url = "https://media-and-learning.eu/type/featured-articles/misinformation-in-virtual-reality/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
print(response.status_code)
print(soup.prettify())
```

```python
#find the div tag which houses the content of the p tag that needs to be scraped
summary = soup.find("div",  attrs = {'class': 'entry-content'})
summary = summary.find_all('p')
```

```python
#get the text from the p tags and store them in a list
sentence_list = []
for sentence in summary:
  text = sentence.get_text(strip = True)
  sentence_list.append(text)

```

```python
#convert the list to a dataframe
df_sentence = pd.DataFrame(sentence_list, columns = ['Text'])
df_sentence
```

```python
#drop irrelevent information
df_sentence.drop([0,7,8,9,10], inplace = True )
df_sentence
```

```python
#store to a csv file
df_sentence.to_csv("Dataset1")
```
