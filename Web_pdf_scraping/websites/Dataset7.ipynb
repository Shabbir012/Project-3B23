```python
#import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#input the link to the website to extract its contents
url = "https://www.hypergridbusiness.com/2023/04/teens-slow-to-adopt-vr-and-more-bad-news-for-the-metaverse/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
#check if the extraction was successful
print(response.status_code)
print(soup.prettify())
```

```python
#find the div tag which houses the content of the p tag that needs to be scraped
summary = soup.find("div", attrs = {'class': "entry-the-content"})
summary = summary.find_all('p')
summary
```

```python
#get the text from thr p tags and store them in a list
sentence_list = []
for sentence in summary:
  text = sentence.get_text(strip = True)
  sentence_list.append(text)
  print(text)

len(sentence_list)
```

```python
#convert list to a dataframe
df = pd.DataFrame(sentence_list, columns = ['Text'])
df.head(5)
```

```python
#convert to csv
df.to_csv("Dataset7")
```
