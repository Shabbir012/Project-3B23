```python
#import the librarie
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#input the link to the website to extract its contents
url = "https://www.theguardian.com/games/2022/jan/25/ive-seen-the-metaverse-and-i-dont-want-it"
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36"}
response = requests.get(url, headers)
soup = BeautifulSoup(response.text, "html5lib")
#check if the extraction was success
print(response.status_code)
print(soup.prettify())
```

```python
#find the p tags that needs to be scraped
summary = soup.find_all("p",  attrs = {'class': 'dcr-hm5hhe'})
summary
```

```python
#get the text from thr p tags and store them in a list
sentence_list = []
for sentence in summary:
  text = sentence.get_text(strip = True)
  sentence_list.append(text)
  print(text)

len(sentence_list)
```

```python
#convert list to a dataframe
df = pd.DataFrame(sentence_list, columns = ['Text'])
df.head(5)
```

```python
#store in a csv file
df.to_csv("Dataset8")
```
