```python
#import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#input the link to the website to extract its contents
url = "https://www.mckinsey.com/about-us/new-at-mckinsey-blog/meet-the-metaverse-creating-real-value-in-a-virtual-world"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
print(response.status_code)
print(soup.prettify())
```

```python
#find the div tag which houses the content of the p tag that needs to be scraped
summary = soup.find("div",  attrs = {'class': 'mdc-o-content-body'})
summary = summary.find_all("p")
summary
```

```python
#get the text from the p tags and store them in a list
sentence_list = []
for sentence in summary:
  text = sentence.get_text(strip = True)
  text = text.replace("Eric:","")
  text = text.replace("Lareina:","")
  text = text.replace("June 15, 2022","")
  sentence_list.append(text)
  print(text)

len(sentence_list)
```

```python
#convert the list to a dataframe
df = pd.DataFrame(sentence_list, columns = ['Text'])
df.head(5)
```

```python
#store to a csv file
df.to_csv("Dataset17")
```
