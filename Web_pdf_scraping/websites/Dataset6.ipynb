```python
#importing libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#input the link to the website to extract its contents
url = "https://www.gamedaily.biz/manufactured-backlash-or-realistic-appraisal/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
#check the status code, to know if the scraping was suceesfuk or not
print(response.status_code)
print(soup.prettify())
```

```python
#find the div tag which houses the content of the p tag that needs to be scraped
div = soup.find("div",  attrs = {'class': 'post_post'})
div
```

```python
paragraph = div.find_all("p")
paragraph
```

```python
#get the text from thr p tags and store them in a list
sentence_list = []
for sentence in paragraph:
  text = sentence.get_text(strip = True)
  sentence_list.append(text)


len(sentence_list)
```

```python
#convert the list to a datafrsme
df = pd.DataFrame(sentence_list, columns = ['Text'])
df.head(5)
```

```python
#store in a  csv file
df.to_csv("Dataset6")
```
