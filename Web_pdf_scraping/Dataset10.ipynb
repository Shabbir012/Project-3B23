```python
#import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#input the link to the website to extract its contents
url = "https://www.theguardian.com/technology/2023/may/21/mark-zuckerbergs-metaverse-vision-is-over-can-apple-save-it"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
print(response.status_code)
print(soup.prettify())
```

```python
#find the div tag which houses the content of the p tag that needs to be scraped
paragraph = soup.find_all("p",  attrs = {'class': 'dcr-4cudl2'})
paragraph
```

```python
#get the text from thr p tags and store them in a list
sentence_list = []
for sentence in paragraph:
  text = sentence.get_text(strip = True)
  sentence_list.append(text)
  print(text)

len(sentence_list)
```

```python
#convert list to dataframe
df = pd.DataFrame(sentence_list, columns = ['Text'])
df.head(5)
```

```python
#convert csv
df.to_csv("Dataset10")
```
