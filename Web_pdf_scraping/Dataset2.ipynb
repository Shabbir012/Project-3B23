```python
#import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#input the link to the website to extract its contents
url = "https://www.theguardian.com/technology/2017/oct/28/virtual-reality-headset-children-cognitive-problems"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
print(response.status_code)
print(soup.prettify())
```

```python
#find the p tags that needs to be scraped
summary = soup.find_all("p",  attrs = {'class': 'dcr-4cudl2'})
summary
```

```python
#get the text from the p tags and store them in a list
sentence_list = []
for sentence in summary:
  text = sentence.get_text(strip = True)
  sentence_list.append(text)
  print(text)

len(sentence_list)
```

```python
#convert the list to a dataframe
df = pd.DataFrame(sentence_list, columns = ['Text'])
df.head(5)
```

```python
#store to a csv file
df.to_csv("Dataset2")
```
